# <font color='blue'>***NORTHWIND AYLIK HASILAT ANALİZİ***</font>

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Northwind_Aylık_Hasılat_Analizi.csv')
```

```python
df= df.rename({'satis_adedi': 'ürün_adedi'}, axis=1)
```

```python
df.head()
```

<font color='red'>***HANGİ AYDA DAHA FAZLA SİPARİŞ VERİLMİŞTİR?***</font>

```python
plt.plot(df['aylar'], df['siparis_sayisi'])
plt.xlabel('Aylar')
plt.ylabel('Sipariş Sayısı')
plt.title('Aylara Göre Sipariş Sayıları')
plt.show()
```

<font color='orange'>***HANGİ AYDA ALINAN ÜRÜN SAYISI EN FAZLADIR?***</font>

```python
plt.plot(df['aylar'], df['ürün_adedi'],color='orange')
plt.xlabel('Aylar')
plt.ylabel('Ürün Adedi')
plt.title('Aylara Göre Ürün Adedi')
plt.show()
```

<font color='green'>***HANGİ AYDA ELDE EDİLEN GELİR EN FAZLADIR?***</font>

```python
plt.plot(df['aylar'], df['toplam_gelir'],color='red')
plt.xlabel('Aylar')
plt.ylabel('Toplam Gelir')
plt.title('Aylara Göre Toplam Gelir')
plt.show()
```

<font color='blue'>***TOPLAM GELİRİN AYLARA GÖRE DAĞILIMI NASILDIR?***</font>

```python
plt.pie(df['siparis_sayisi'], labels=df['aylar'], autopct='%1.1f%%')
plt.title('Toplam Gelirin Aylara Göre Dağılımı')
plt.show()
```
